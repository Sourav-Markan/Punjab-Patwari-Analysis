```python
import csv
import pandas as pd

def RepresentsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
```

```python
# opening the CSV file
with open('result.csv', mode ='r')as file:
    csvFile = csv.reader(file)
    index = 0
    max_number = float('0.0')
  
    result = []

    temp = 0
    for line in csvFile:
        temp += 1
        array = line[0].split()
        if len(array) > 0 and RepresentsInt(array[0]):
            category = ''
            name = ''
            Gender = ''
            for field in array[2:]:
                if field == 'F' or field == 'M':
                    Gender = field
                    break
                name = name + field + ' '
        
            for field in reversed(array[:-1]):
                category = category + field + ' '
                if field == 'F' or field == 'M':
                    break
            rev_category = category.strip().split()[:-2]
            
            category = '' 
            for field in reversed(rev_category):
                category = category + field + ' '

            category = category.strip()
            data = [array[0], array[1], name, Gender ,category, float(array[-1])]
            result.append(data)  

    df = pd.DataFrame(result, columns=['Sr', 'Roll', 'Name', 'Gender', 'Category', 'Number'])
    df_sorted_number = df.sort_values('Number', ascending=False)
    
    result = []
    rank = 0
    for index, row in df_sorted_number.iterrows(): 
        rank += 1
        data = [row['Sr'], row['Roll'], row['Name'], row['Gender'], row['Category'], row['Number'], rank]
        result.append(data) 

    df = pd.DataFrame(result, columns=['Sr', 'Roll', 'Name', 'Gender', 'Category', 'Number', 'GRank'])
    
    group = df.groupby('Category')
    result = []
    for name, grp in group:
        print (name)
        rank = 0
        df_sorted_grp = grp.sort_values('Number', ascending=False)
        for index, row in df_sorted_grp.iterrows():
            rank += 1
            data = [row['Roll'], row['Name'], row['Gender'], row['Category'], row['Number'], row['GRank'], rank]
            result.append(data) 
        
    df = pd.DataFrame(result, columns=['Roll', 'Name', 'Gender','Category', 'Number', 'GRank', 'CRank'])    
    
```

```python
df.sort_values('GRank').to_csv('out.csv')
```
